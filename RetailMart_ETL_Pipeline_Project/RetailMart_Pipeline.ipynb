```python
import pandas as pd
import sqlite3

def run_pipeline():
    try:
        # Load CSV files:

        sales_df = pd.read_csv('sales_data.csv')
        products_df = pd.read_csv('products.csv')
        stores_df = pd.read_csv('stores.csv')

        # Clean data:

        sales_df = sales_df.drop_duplicates()
        sales_df['quantity'] = pd.to_numeric(
            sales_df['quantity'],
            errors='coerce'
        ).fillna(0)

        sales_df['amount'] = pd.to_numeric(
            sales_df['amount'],
            errors='coerce'
        )

        sales_df = sales_df.dropna(subset=['amount'])

        sales_df['sale_date'] = pd.to_datetime(
            sales_df['sale_date'],
            errors='coerce'
        )

        products_df['price'] = pd.to_numeric(
            products_df['price'],
            errors='coerce'
        ).fillna(0)

        # Merge DataFrames:

        final_df = sales_df.merge(
            products_df,
            on='product_id',
            how='left'
        ).merge(
            stores_df,
            on='store_id',
            how='left'
        )

        # Transform data:

        final_df['total_revenue'] = (
            final_df['quantity'] *
            final_df['price']
        )

        # Load into SQLite database:

        conn = sqlite3.connect('retailmart.db')

        final_df.to_sql(
            'retail_sales',
            conn,
            if_exists='replace',
            index=False
        )

        conn.close()

        print("Pipeline completed successfully!")
        return final_df

    except FileNotFoundError:
        print("Error: One or more CSV files are missing. Please check the file names and paths.")
```

```python
final_df = run_pipeline()
```
